```python
import pandas as pd
from pathlib import Path
import urllib.request

# ----------------------------
# INPUT
# ----------------------------
URL = "https://github.com/Raziye-Aghapour/ReEDS_Morris/raw/refs/heads/main/Fuel%20price%20files/dollaryear.csv"

START_IDX = 0
END_IDX = 200  # inclusive => 000..200

OUT_FILE = "dollaryear_updated.csv"

# ----------------------------
# READ ORIGINAL
# ----------------------------
df = pd.read_csv(URL)

# ----------------------------
# BUILD NEW SCENARIOS
# (scenario names used by ReEDS, not the alpha_/ng_demand_ filenames)
# ----------------------------
new_scenarios = [f"AEO_2025_West_South_Central_morris_{i:03d}" for i in range(START_IDX, END_IDX + 1)]
new_df = pd.DataFrame({"Scenario": new_scenarios, "Dollar.Year": [2024] * len(new_scenarios)})

# ----------------------------
# APPEND, KEEP ORIGINAL IF DUPLICATE
# ----------------------------
df_out = pd.concat([df, new_df], ignore_index=True)
df_out = df_out.drop_duplicates(subset=["Scenario"], keep="first")

# (optional) sort for readability
def sort_key(s):
    s = str(s)
    if "AEO_2025_West_South_Central_morris_" in s:
        return (1, int(s.split("_")[-1]))
    return (0, s.lower())

df_out = df_out.sort_values(by="Scenario", key=lambda col: col.map(sort_key)).reset_index(drop=True)

# ----------------------------
# SAVE
# ----------------------------
df_out.to_csv(OUT_FILE, index=False)
print("Wrote:", Path(OUT_FILE).resolve())
print("Rows:", len(df_out))
```
