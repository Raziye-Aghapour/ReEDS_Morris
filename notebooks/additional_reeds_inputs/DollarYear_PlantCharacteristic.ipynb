```python
import pandas as pd
import re

URL = "https://raw.githubusercontent.com/Raziye-Aghapour/ReEDS_Morris/refs/heads/main/Plant%20characteristics%20files/dollaryear.csv"

START_IDX = 0
END_IDX = 200  # inclusive => 000..200

df = pd.read_csv(URL)
df["Dollar.Year"] = df["Dollar.Year"].astype(int)

# Find all rows ending with _morris_000 and expand
new_rows = []
pattern = re.compile(r"^(?P<prefix>.+)_morris_000$")

for _, row in df.iterrows():
    scen = str(row["Scenario"])
    m = pattern.match(scen)
    if not m:
        continue
    prefix = m.group("prefix")
    year = int(row["Dollar.Year"])
    for i in range(START_IDX, END_IDX + 1):
        new_rows.append({"Scenario": f"{prefix}_morris_{i:03d}", "Dollar.Year": year})

df_out = pd.concat([df, pd.DataFrame(new_rows)], ignore_index=True)

# Keep original if duplicates exist (e.g., _000 already present)
df_out = df_out.drop_duplicates(subset=["Scenario"], keep="first")

OUT_FILE = "dollaryear_morris_000_to_200.csv"
df_out.to_csv(OUT_FILE, index=False)

print("Wrote:", OUT_FILE)
print("Rows:", len(df_out))
```
